# 04. Merge the dataframe with your df_ords dataframe
# Use the indicator argument to check whether there was a full match between the two dataframes

```python
df_merged_large = df_ords.merge(df_ords_prior, on = 'order_id', indicator = True)
```

```python
# Check output of merged dataframe
df_merged_large.head()
```

```python
# Re-import the dataset (adjust the path and file name as needed)
df_ords_original = pd.read_csv(os.path.join(path, '02 Data', 'Original Data', 'orders.csv'))

# Check the first few rows to verify if 'order_id' and 'user_id' are different
df_ords_original.head()
```

```python
df_merged_large.head()
```

```python
# Perform the merge
# This joins df_merged_large with df_ords_original based on 'order_id'
# The aim is to add the correct 'user_id' from df_ords_original to df_merged_large
df_merged_corrected = df_merged_large.merge(df_ords_original[['order_id', 'user_id']], on='order_id', how='left')

# After merging, df_merged_corrected will have two user_id columns: 
# 'user_id_x' (from df_merged_large) and 'user_id_y' (from df_ords_original)
# Replace 'user_id_x' with the values from 'user_id_y'
df_merged_corrected['user_id'] = df_merged_corrected['user_id_y']

# Drop the old/unnecessary user_id columns
df_merged_corrected.drop(columns=['user_id_x', 'user_id_y'], inplace=True)

# Reorder the columns
df_merged_corrected = df_merged_corrected[['order_id', 'user_id', 'order_number', 'orders_day_of_week', 'order_hour_of_day', 'days_since_prior_order', 'product_id', 'add_to_cart_order', 'reordered', '_merge']]
```
